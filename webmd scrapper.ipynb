```python
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}
request = requests.get('https://www.webmd.com/drugs/2/index',headers=headers)
soup= BeautifulSoup(request.text, 'html.parser')
```

```python
main_Dict={}
for i in soup.find_all('ul',{"class": "browse-letters squares"})[0].find_all('a'):
    x=f'{i.get_text().strip()}'
    main_Dict[x]={}
    main_Dict[x]['link']=f"https://www.webmd.com{i.attrs['href']}"
```

```python
for key in main_Dict.keys():
  request = requests.get(main_Dict[key]['link'],headers=headers)
  soup= BeautifulSoup(request.text, 'html.parser')
  for k in soup.find_all('ul',{"class": "browse-letters squares sub-alpha"})[0].find_all('a'):
    main_Dict[key][f'{k.get_text().strip()}']={}
    main_Dict[key][f'{k.get_text().strip()}']['link']=f"https://www.webmd.com{k.attrs['href']}"
```

```python
for key1 in main_Dict.keys():
  for key2 in main_Dict[key1]:
    if key2 !='link':
      request = requests.get(main_Dict[key1][key2]['link'],headers=headers)
      soup= BeautifulSoup(request.text, 'html.parser')
      for k in soup.find_all('a',{"class": "alpha-drug-name"}):
        main_Dict[key1][key2][f'{k.get_text().strip()}']={}
        main_Dict[key1][key2][f'{k.get_text().strip()}']['link']=f"https://www.webmd.com{k.attrs['href']}"
```

```python
fu = ['monograph-content monograph-content-holder','how-to-use-section', 'side-effects-container center-content','precautions-container','interactions-container center-content']
for key1 in main_Dict.keys():
  for key2 in main_Dict[key1].keys():
    if key2!='link':
      for key3 in main_Dict[key1][key2].keys():
        if key3!='link':
          request = requests.get(main_Dict[key1][key2][key3]['link'],headers=headers)
          soup= BeautifulSoup(request.text, 'html.parser')
          for a in fu:
            if a=='monograph-content monograph-content-holder':
                  for i in soup.find_all('div',{"class": f"{a}"}):
                    main_Dict[key1][key2][key3]['Uses']=' '.join([item.get_text() for item in list(i.find_all('p'))])
            if a=='how-to-use-section':
                  for i in soup.find_all('div',{"class": f"{a}"}):
                    main_Dict[key1][key2][key3]['How to use']=' '.join([item.get_text() for item in list(i.find_all('p'))])

            if a=='side-effects-container center-content':
                  for i in soup.find_all('div',{"class": f"{a}"}):
                      main_Dict[key1][key2][key3]['Side Effects']=' '.join([item.get_text() for item in list(i.find_all('p'))])
            if a=='precautions-container':
                  for i in soup.find_all('div',{"class": f"{a}"}):
                    main_Dict[key1][key2][key3]['Precautions']=' '.join([item.get_text() for item in list(i.find_all('p'))])
            if a=='interactions-container center-content':
                  for i in soup.find_all('div',{"class": f"{a}"}):
                    main_Dict[key1][key2][key3]['Interactions']=' '.join([item.get_text() for item in list(i.find_all('p'))])
```
